```python
def heapify_verbose(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    print(f"\n[heapify 시작] i={i}, arr={arr}")

    if left < n:
        print(f"  왼쪽 자식 index={left}, 값={arr[left]}")
    if right < n:
        print(f"  오른쪽 자식 index={right}, 값={arr[right]}")

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        print(f"  교환: arr[{i}]={arr[i]} <-> arr[{largest}]={arr[largest]}")
        arr[i], arr[largest] = arr[largest], arr[i]
        print(f"  교환 후: {arr}")
        heapify_verbose(arr, n, largest)
    else:
        print("  교환 없음")


def build_max_heap_verbose(arr):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        print(f"\n===== i = {i} 수선 시작 =====")
        heapify_verbose(arr, n, i)

    print("\n최종 최대 힙:", arr)


data = [3, 5, 1, 10, 2, 7, 6, 4, 8]
build_max_heap_verbose(data)
```
